```python
from typing_extensions import Annotated, TypedDict
import operator
```

```python
from langchain_core.messages import AnyMessage, HumanMessage, AIMessage
```

```python
class GraphState(TypedDict):
    messages: Annotated[list[AnyMessage], operator.add]
```

```python
def llm(state:GraphState)->dict:
    """call the LLM using conversation messages and append AI messages"""
    response = chat_model.invoke(state["messages"])
    return {
       "messages":[response]
   }
```

```python
def token_counter(state:GraphState)->dict:
    """count tokens(simple word count) in last AI message"""
    last_ai_message = state["messages"][-1]
    text=last_ai_message.content
    token_count = len(text.split())
    summary=f"Token number in the generated answer (word count) is {token_count}"
    return {
        "messages":[AIMessage(content=summary)]
    }
```
